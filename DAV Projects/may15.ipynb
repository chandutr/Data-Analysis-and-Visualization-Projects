```python
import pandas as pd
temp_df=pd.DataFrame({
    'City':['Newyork','Chicago','Orlando','SanFrancisco'],
    'Temperature':[21,14,35,25]
})
temp_df
```

```python
windspeed_df=pd.DataFrame({
    'City':['Chicago','Newyork','Baltimore'],
    'Windspeed':[6,7,2]
})
windspeed_df
```

```python
pd.merge(temp_df,windspeed_df, on='City')#by default - inner join
```

```python
pd.merge(temp_df,windspeed_df, on='City', how="left")
```

```python
pd.merge(temp_df,windspeed_df, on='City', how="right")
```

```python
pd.merge(temp_df,windspeed_df, on='City', how="outer")
```

```python
pd.concat([temp_df,windspeed_df])
```

```python
pd.concat([temp_df,windspeed_df], axis=1)
```
