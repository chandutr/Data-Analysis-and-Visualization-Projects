```python
df['Sale Date']=pd.to_datetime(df['Sale Date'])
df.dtypes
```

Using apply method
df['Month']=df['Sale Date'].apply(lambda x:x.month)

```python
df['Month']=df['Sale Date'].dt.month
```

```python
df[['Sale Date','Month']].head(10)
```

```python
df.columns
```

```python
df['Month'].value_counts()
```

```python
month_sales=df['Month'].value_counts()
month_sales
```

```python
month_sales.sort_index(inplace=True)
month_sales
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,5))

plt.plot(month_sales, linestyle= 'dotted',color='red', marker='*')

plt.xticks(range(1,13))
plt.xlabel('Month')
plt.ylabel('Total No of Sales')
plt.title('Monthly Sales Volume')
```

Comparing the daily sales volume of diff property types in the month of july(you have to create multiple line plot(for diff property types) in a single graph)

```python
df['Property Type'].value_counts()
```

```python
df=df[df['Property Type']!='O']
```

```python
df['Property Type'].value_counts()
```

```python
df.describe()
```

```python
jul=df[df['Month']==7]
jul.head()
```

```python
daily_sales=jul.groupby(['Sale Date','Property Type'])['Property Type'].count().unstack()
daily_sales
```

```python
plt.rcParams['figure.figsize']=(18,5)
daily_sales.plot()
plt.xlabel('Day')
plt.ylabel('Total No of Sales')
plt.title('Daily Sales Volume in July 2018')
```

Average house prices for diff properties in the month july 2018 (box plot and bar plot)

```python
july=df[df['Month']==7]
july.head()
```

```python
avg_house_price=july.groupby(['Sale Date','Property Type'])['Price'].mean().unstack()
avg_house_price
```

A box plot gives a five-number summary of a set of data which is-
Minimum – It is the minimum value in the dataset excluding the outliers.
First Quartile (Q1) – 25% of the data lies below the First (lower) Quartile.
Median (Q2) – It is the mid-point of the dataset. Half of the values lie below it and half above.
Third Quartile (Q3) – 75% of the data lies below the Third (Upper) Quartile.
Maximum – It is the maximum value in the dataset excluding the outliers.

```python
avg_house_price.plot(kind='box')
```

Obtain the average house price distribution for diff property types in the month september 2018( box plot x-axis=property types, y-axis=average property price in September)

```python
sep=df[df['Month']==9]
sep.head()
```
